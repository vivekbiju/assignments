#One Way Anova

### Case Study 1

A researcher recruits 30 students to participate in a study. The students are randomly assigned to use one of three studying techniques for the next three weeks to prepare for an exam. At the end of the three weeks, all of the students take the same test.
group1 = [85, 86, 88, 75, 78, 94, 98, 79, 71, 80]
group2 = [91, 92, 93, 85, 87, 84, 82, 88, 95, 96]
group3 = [79, 78, 88, 94, 92, 85, 83, 85, 82, 81]

```python
group1 = [85, 86, 88, 75, 78, 94, 98, 79, 71, 80]
group2 = [91, 92, 93, 85, 87, 84, 82, 88, 95, 96]
group3 = [79, 78, 88, 94, 92, 85, 83, 85, 82, 81]
```

#### Method 1

```python
#importing f_oneway from scipy.stats
from scipy.stats import f_oneway
f_stat, p_value = f_oneway(group1, group2,group3)
```

```python

if p_value<0.05:
  print("Reject H0")
else:
  print("Accept H0")
```

We accept the H0, ie study technique has no impact on marks.

### Case Study 3

```python
plant_data = pd.read_csv("/content/PlantGrowth.csv")
plant_data
```

```python
plant_data.group.unique()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# weight dependent group independent
model = ols('weight ~ group', data=plant_data).fit()

anova_tab = sm.stats.anova_lm(model, typ=1)

anova_tab
```

```python
if anova_tab.loc['group']['PR(>F)'] < 0.05:
  print("Reject H0")
else:
  print("Accept H0")
```

# Two Way Anova

```python
df = pd.read_csv("/content/Anova2.csv")
df
```

```python
# melting check again
df_melt = pd.melt(df, id_vars=["Genotype"], value_vars=["1_year","2_year","3_year"])
df_melt
```

```python
df_melt.columns=['Genotype','Years','Growth_Value']
```

```python
df_melt.head()
```

```python
# Growth value dependent
model = ols('Growth_Value ~ C(Genotype) + C(Years) + C(Genotype):C(Years)', data = df_melt).fit()

anova_table = sm.stats.anova_lm(model, type=2)

anova_table
```

```python
# since all probability vales are < 0.05 reject H0 ie the 2 factors as well as combined effect is applicable
```
